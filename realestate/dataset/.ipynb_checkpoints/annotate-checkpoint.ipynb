```python
import pandas as pd
```

```python
src=pd.read_excel('sourcedata.xlsx')
src.columns=src.columns.str.replace("/tuple/",'')
src
```

```python
target_agent=pd.read_excel('target_agent.xlsx')
target_house=pd.read_excel('house_details_target.xlsx')
target_agent.columns=target_agent.columns.str.replace("/tuple/","")
target_house.columns=target_house.columns.str.replace("/tuple/","")
target_agent
```

```python
target_house
```

```python
source_final=[]
target_final=[]
for source in src.columns:
    for target1 in target_house.columns:
        source_final.append(source)
        target_final.append(target1)
    for target2 in target_agent.columns:
        source_final.append(source)
        target_final.append(target2)
match=pd.DataFrame({'source':source_final,'target':target_final})
# match.to_excel('annotate.xlsx')
```
