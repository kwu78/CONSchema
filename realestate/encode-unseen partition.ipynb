```python
import pandas as pd
```

```python
train=pd.read_excel('./dataset/unseen_split/train_real_estates.xlsx')
train=train.loc[:,~train.columns.str.match("Unnamed")]
train
```

```python
source_constraint=pd.read_excel("./dataset/source_constraint.xlsx")
house_constraint=pd.read_excel('./dataset/house_constraint.xlsx')
agent_constraint=pd.read_excel('./dataset/agent_constraint.xlsx')
target_constraint=house_constraint.append(agent_constraint)
target_constraint=target_constraint.loc[:,~target_constraint.columns.str.match("Unnamed")]
target_constraint
```

```python
target_constraint=target_constraint.drop_duplicates()
```

```python
real_estate=pd.merge(train,source_constraint,on='source',how='left')

real_estate=real_estate.loc[:,~real_estate.columns.str.match("Unnamed")]


    
```

```python
real_estate
```

```python
real_estate=pd.merge(real_estate,target_constraint,on='target',how='left')
```

```python
real_estate=real_estate.drop(columns=['source','target'])
```

```python
type1=pd.get_dummies(real_estate['type_x'])
type1=type1.add_suffix('_x')
type2=pd.get_dummies(real_estate['type_y'])
type2
type2=type2.add_suffix('_y')
type12=pd.concat([type1,type2],axis=1,join='inner')
type12
```

```python
import json 
```

```python
f=open('./dataset/unseen_split/estate_outputs/4_train_raw_outputs.json',)
real_estate_raw = json.load(f)
real_estate_raw.keys()
for b in range(5):
    a=real_estate_raw['batch_'+str(b)]
    if not b==0:
        real_estate_raw['batch_0'].extend(a)
len(real_estate_raw['batch_0'])
import torch
import torch.nn as nn
tversion=torch.tensor(real_estate_raw['batch_0'])
m=torch.nn.Softmax(dim=1)
output_realestate_train=[i[1] for i in m(tversion).tolist()]
len(output_realestate_train)
```

```python
real_estate=pd.concat([real_estate,type12],axis=1,join='inner')
real_estate=real_estate.drop(columns=['type_x','type_y'])
real_estate['outputscore']=output_realestate_train
real_estate['label'].to_csv('./dataset/unseen_split/4train_label.csv')
real_estate=real_estate.drop(columns=['label'])
```

```python
real_estate.to_csv('./dataset/unseen_split/4train_encode.csv')
```
