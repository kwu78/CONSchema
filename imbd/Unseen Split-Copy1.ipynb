```python
label=match.iloc[0:,3]
input1=label.groupby(label.index).sum()

input2=input1.head(10)
print (input1)
```

```python
freq={}
for k,v in input1.iteritems():
    if v in freq:
        freq[v].append(k)
    else:
        freq[v]=[k]

freq
```

```python
matches=pd.read_excel("./unseen_split/imdb.xlsx")
totalones=len(matches[matches['label']==1])
totalCols=len(input1)
print (totalones)
print (totalCols)
```

```python
import numpy as np

np.random.choice(freq[1])
```

```python
nVal=int(totalones*0.1)
nValCols=int(0.1*totalCols)
val=[]
print(nVal)
while nVal>0 and len(val)<nValCols:
    newChoice=np.random.choice(nVal+1,1)[0]
    if int(newChoice) in freq:
        if len(freq[newChoice])>0:
            print ("newChoiceselected")
            print (newChoice)
            decision=np.random.choice(freq[newChoice])
            val.append(decision)
            freq[newChoice].remove(decision)
            nVal=nVal-newChoice
while len(val)<nValCols:
    decision=np.random.choice(freq[0])
    val.append(decision)
    freq[0].remove(decision)
print (val)
```

```python
nTest=int(totalones*0.2)
nTestCols=int(0.2*totalCols)
test=[]
while nTest>0 and len(test)<nTestCols:
    newChoice=np.random.choice(nTest+1,1)[0]
    if int(newChoice) in freq:
        if len(freq[newChoice])>0:
            print ("newChoiceselected")
            print (newChoice)
            decision=np.random.choice(freq[newChoice])
            test.append(decision)
            freq[newChoice].remove(decision)
            nTest=nTest-newChoice
while len(test)<nTestCols:
    decision=np.random.choice(freq[0])
    test.append(decision)
    freq[0].remove(decision)
print (test)
```

```python
train=[]
for i in freq:
    if len(freq[i])>0:
        for item in freq[i]:
            train.append(item)
train
```

```python
import xlsxwriter
training=pd.DataFrame()
writer1 = pd.ExcelWriter('./unseen_split/train_imdb.xlsx', engine='xlsxwriter')
for m in train:
    training=training.append(matches.loc[matches['imdb']==m])
training.to_excel(writer1,sheet_name='imdb_train',index=False)
print (len(training))
writer1.save()
training
```
