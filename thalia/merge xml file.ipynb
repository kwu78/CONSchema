```python
import re
import pandas as pd
import os, os.path, sys
import glob
pattern=r"(?<=\.\/goldenmapping\/)\w+"
pattern2=r"(?<=\._)"
source=[]
target=[]
size1=[]
type1=[]
size2=[]
type2=[]
label=[]
def run2(files):
    csv_files = glob.glob(files +"/*.csv")
    for f in csv_files:
        golden_mapping=pd.read_csv(f,header=None)
        temp_source=[]
        temp_target=[]
        m=re.search(pattern,f).group(0)
        schools=m.split('2')
#         print(pd.read_xml(schools[0]+".xml",encoding='cp1252').columns)
        university1=pd.read_excel('./columns/'+schools[0]+".xlsx")
        university2=pd.read_excel('./columns/'+schools[1]+".xlsx")
        for s in pd.read_excel("./xlsx/"+schools[0]+".xlsx").columns:
            temp_source.append(schools[0]+"."+s)
#             if schools[0]=='fsu':
#                 print(university1[university1['column']==s])

        for t in pd.read_excel("./xlsx/"+schools[1]+".xlsx").columns:
            temp_target.append(schools[1]+"."+t)

#         print(temp_source)
#         print(temp_target)
        for temp in temp_source:
            for temp2 in temp_target:
                match1=temp.replace("/",'.')
                match2=temp2.replace("/",'.')
                match1=match1.replace('.', '', 1)
                match2=match2.replace('.', '', 1)

                source.append(temp)
                target.append(temp2)
#                 print(golden_mapping[0])
#                 print(match2)
#                 print(golden_mapping[0])
                if (golden_mapping[0].eq(match1)).any():
                    if golden_mapping[golden_mapping[0]==match1][1].iloc[0]==match2:
                        label.append(1)
                    else:
                        label.append(0)
                else:
                    label.append(0)
#                     print(golden_mapping)
#                 print(university1[university1['column']==temp.split("_")[1]]['size'].iloc[0])
                size1.append(university1[university1['column']==temp.split(".")[1]]['size'].iloc[0])
                type1.append(university1[university1['column']==temp.split(".")[1]]['type'].iloc[0])
                size2.append(university2[university2['column']==temp2.split(".")[1]]['size'].iloc[0])
                type2.append(university2[university2['column']==temp2.split(".")[1]]['type'].iloc[0])
run2("./goldenmapping")
# print(source)
# print(target)
```
