```python
import pandas as pd
```

```python
train=pd.read_excel('./train_thalia.xlsx')
train=train.loc[:,~train.columns.str.match("Unnamed")]
train
```

```python
train=train.drop(columns=['source','target'])
train
```

```python
type1=pd.get_dummies(train['type1'])
type1=type1.add_suffix('_x')
type2=pd.get_dummies(train['type2'])
type2
type2=type2.add_suffix('_y')
type12=pd.concat([type1,type2],axis=1,join='inner')
type12
```

```python
import json 
```

```python
f=open('./thalia_kevin/4_train_raw_outputs.json',)
train_raw = json.load(f)
train_raw.keys()
for b in range(3):
    a=train_raw['batch_'+str(b)]
    if not b==0:
        train_raw['batch_0'].extend(a)
len(train_raw['batch_0'])
import torch
import torch.nn as nn
tversion=torch.tensor(train_raw['batch_0'])
m=torch.nn.Softmax(dim=1)
output_thalia_train=[i[1] for i in m(tversion).tolist()]
len(output_thalia_train)
```

```python
train=pd.concat([train,type12],axis=1,join='inner')
train=train.drop(columns=['type1','type2'])
train['outputscore']=output_thalia_train
train['label'].to_csv('./4train_label.csv')
train=train.drop(columns=['label'])
```

```python
train.to_csv('./4train_encode.csv')
```
